```python
from sklearn.tree import DecisionTreeClassifier
from sklearn.metrics import roc_auc_score,roc_curve,auc,classification_report
from sklearn.preprocessing import LabelEncoder,OrdinalEncoder
from sklearn.model_selection import train_test_split
```

```python
df[['Vehicle_Age']].value_counts()
```

```python
categories=[['< 1 Year', '1-2 Year', '> 2 Years'] ]
encoder = OrdinalEncoder(categories=categories)
df['Vehicle_Age'] = encoder.fit_transform(df[['Vehicle_Age']])
```

```python
df.head()
```

```python
labelEn=LabelEncoder()
df['Gender']=labelEn.fit_transform(df[['Gender']])
df['Vehicle_Damage']=labelEn.fit_transform(df[['Vehicle_Damage']])
```

```python
df.head()
```

```python
X=df.drop(columns=['Response','id'])
Y=df['Response']
```

```python
X_train, X_test, y_train, y_test = train_test_split(X, Y, test_size=0.2, random_state=42)
```

```python
df.to_csv("data/cleaned_data.csv",index=False)
```

#### Using the Decision Tree Classifier

```python
dt=DecisionTreeClassifier(criterion='gini',      
                            max_depth=13,        
                            min_samples_split=5, 
                            min_samples_leaf=5,  
                            max_features='sqrt', 
                            class_weight='balanced',
                            random_state=42)  
dt.fit(X_train,y_train)
```

#### Classification Report 

```python
y_pred=dt.predict(X_test)
print(classification_report(y_test, y_pred))
```

#### Plotting the Receiver Operator Characteristic Curve for Understanding the Model Performance

```python
y_probs = dt.predict_proba(X_test)[:, 1] 
```

```python
# Computing the ROC CURVE AND AUC SCORE 
fpr, tpr, thresholds = roc_curve(y_test, y_probs)
roc_auc = auc(fpr, tpr)
```

```python
# PLotting ROC Curve 
plt.figure(figsize=(8, 6))
plt.plot(fpr, tpr, color='blue', lw=2, label=f'ROC Curve (AUC = {roc_auc:.2f})')
plt.plot([0, 1], [0, 1], color='grey', linestyle='--')  # Random classifier
plt.xlabel('False Positive Rate')
plt.ylabel('True Positive Rate')
plt.title('ROC Curve for Decision Tree Classifier')
plt.legend(loc='lower right')
plt.show()

# Print AUC Score
print(f"ROC-AUC Score: {roc_auc:.4f}")
```

#### Removing the Feature Importance using Decision Trees

```python
feature_importances = pd.Series(dt.feature_importances_, index=X_train.columns)
feature_importances.sort_values(ascending=False).plot(kind='bar', figsize=(10,5), title="Feature Importance")
plt.show()
```
